```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def plot_optimized_histogram(data, column_name):
    """
    Automatically calculates the optimal number of bins using the
    Freedman-Diaconis rule and plots the distribution.
    """
    series = data[column_name].dropna()

    # Calculate IQR (Interquartile Range)
    q25, q75 = np.percentile(series, [25, 75])
    iqr = q75 - q25

    # Calculate bin width: 2 * IQR * n^(-1/3)
    bin_width = 2 * iqr / (len(series) ** (1/3))

    # Calculate number of bins
    if bin_width > 0:
        nbins = int((series.max() - series.min()) / bin_width)
    else:
        nbins = 10 # Fallback

    print(f"Optimal bins detected: {nbins}")

    # Plotting
    plt.figure(figsize=(10, 6))
    sns.histplot(series, bins=nbins, kde=True, color='teal', edgecolor='white')
    plt.title(f'Optimized Histogram: {column_name} (Bins={nbins})')
    plt.xlabel(column_name)
    plt.ylabel('Frequency')
    plt.show()

# Usage Example:
plot_optimized_histogram(housing_data, 'Rooms')
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# Sample Data: Customer spending in three different regions
np.random.seed(42)
data_region_a = np.random.normal(loc=50, scale=10, size=100)
data_region_b = np.random.normal(loc=60, scale=15, size=100)
data_region_c = np.random.normal(loc=45, scale=5, size=100)

# Introduce some outliers
data_region_a = np.append(data_region_a, [10, 100, 105])
data_region_b = np.append(data_region_b, [120, 125, 5])

# Combine into a DataFrame for easier plotting with Seaborn/Pandas
df = pd.DataFrame({
    'Spending': np.concatenate([data_region_a, data_region_b, data_region_c]),
    'Region': ['A']*len(data_region_a) + ['B']*len(data_region_b) + ['C']*len(data_region_c)
})

# Basic Box Plot using Seaborn
plt.figure(figsize=(10, 6))
sns.boxplot(x='Region', y='Spending', data=df, palette='viridis');
plt.title('Customer Spending Distribution by Region')
plt.xlabel('Region')
plt.ylabel('Spending ($)')
plt.show();
```

```python
plt.figure(figsize=(8, 5))
plt.boxplot([data_region_a, data_region_b, data_region_c],
            labels=['Region A', 'Region B', 'Region C'],
            patch_artist=True, # Fills the box with color
            medianprops={'color': 'red', 'linewidth': 2},
            boxprops={'facecolor': 'lightgreen', 'edgecolor': 'darkgreen'},
            whiskerprops={'color': 'blue'},
            capprops={'color': 'black'})
plt.title('Matplotlib Box Plot with Custom Styles')
plt.ylabel('Spending ($)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Create a melted DataFrame for pandas boxplot
df_melted = df.melt(id_vars=['Region'], var_name='Metric', value_name='Value')

plt.figure(figsize=(8, 5))
df.boxplot(column='Spending', by='Region', grid=False, rot=45, figsize=(8,5))
plt.title('Pandas Box Plot (Spending by Region)')
plt.suptitle('') # Suppress default pandas title
plt.ylabel('Spending ($)')
plt.xlabel('Region')
plt.show()
```

```python
# The plot from the beginning of the article
plt.figure(figsize=(10, 6))
sns.boxplot(x='Region', y='Spending', data=df, palette='cubehelix')
plt.title('Customer Spending Distribution by Region (Seaborn)')
plt.xlabel('Region')
plt.ylabel('Spending ($)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Region', y='Spending', data=df, palette='GnBu_r')
sns.stripplot(x='Region', y='Spending', data=df, color='black', size=4, jitter=0.2)
plt.title('Box Plot with Individual Data Points (Jitter)')
plt.xlabel('Region')
plt.ylabel('Spending ($)')
plt.show()
```
