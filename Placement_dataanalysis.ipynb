```python
# Select only numerical columns for correlation calculation
numerical_df = df.select_dtypes(include=np.number)

# Calculate the correlation matrix
correlation_matrix = numerical_df.corr()

# Display the correlation matrix
display(correlation_matrix)
```

Now that we have the correlation matrix, let's find the pairs of columns with the highest absolute correlation. A correlation value close to 1 or -1 indicates a strong positive or negative linear relationship, respectively. A value close to 0 indicates a weak or no linear relationship.

```python
# Unstack the correlation matrix to get pairs of columns and their correlation values
corr_pairs = correlation_matrix.unstack()

# Sort the correlation pairs in descending order of absolute value
sorted_corr_pairs = corr_pairs.sort_values(ascending=False)

# Exclude self-correlations (correlation of a variable with itself is always 1)
# and duplicate pairs (e.g., (A,B) and (B,A) are the same)
# To do this, we'll iterate and keep only unique pairs where the first element is less than the second

strongest_correlations = []
for (idx1, idx2), value in sorted_corr_pairs.items():
    if idx1 != idx2 and abs(value) < 1.0:
        # Ensure we only add each pair once (e.g., 'A'-'B' but not 'B'-'A')
        if (idx2, idx1, value) not in strongest_correlations:
            strongest_correlations.append((idx1, idx2, value))

# Display the top 10 strongest positive and negative correlations
print("Top 10 strongest correlations (absolute value):")
# Filter out duplicates and self-correlations, then sort by absolute value
all_pairs = correlation_matrix.stack().reset_index()
all_pairs.columns = ['Variable1', 'Variable2', 'Correlation']

# Remove self-correlations
all_pairs = all_pairs[all_pairs['Variable1'] != all_pairs['Variable2']]

# Remove duplicate pairs (e.g., (A,B) and (B,A))
all_pairs['Sorted_Pair'] = all_pairs.apply(lambda row: tuple(sorted([row['Variable1'], row['Variable2']])), axis=1)
all_pairs = all_pairs.drop_duplicates(subset=['Sorted_Pair'])

# Sort by absolute correlation value in descending order
strongest_correlations = all_pairs.reindex(all_pairs['Correlation'].abs().sort_values(ascending=False).index)

display(strongest_correlations.head(10))
```

```python
# Filter for 'Not Placed' students
not_placed_students = df[df['status'] == 'Not Placed']

# Get the 'ssc_p' for these students
ssc_p_not_placed = not_placed_students['ssc_p']

# Calculate the 75th percentile of 'ssc_p' for not placed students
percentile_75 = ssc_p_not_placed.quantile(0.75)

# Count how many not placed students have ssc_p less than 65%
count_less_than_65 = ssc_p_not_placed[ssc_p_not_placed < 65].count()

# Total number of not placed students
total_not_placed = len(ssc_p_not_placed)

# Percentage of not placed students with ssc_p less than 65%
percentage_less_than_65 = (count_less_than_65 / total_not_placed) * 100

print(f"75th percentile of Secondary Education percentage for Not Placed students: {percentile_75:.2f}%")
print(f"Number of Not Placed students with ssc_p < 65%: {count_less_than_65}")
print(f"Total Not Placed students: {total_not_placed}")
print(f"Percentage of Not Placed students with ssc_p < 65%: {percentage_less_than_65:.2f}%")
```

```python
import pandas as pd
import numpy  as np


df =pd.read_csv('data.csv')
outlier_columns = ['Ad Spend', 'Profit Margin', 'Revenue']

# Cliping the values of Each Column in outlier column using IQR Method.
for column in outlier_columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)

    iqr = Q3-Q1
    upper_bound = Q3 + 1.5 * iqr
    lower_bound = Q1 - 1.5 * iqr
    df[column] = df[column].clip(lower=lower_bound, upper=upper_bound)

    df[column] =df[column].clip(lower_bound,upper_bound)

```

```python
#identify the columns where the percentage of null values exceed 25%  and store them in a list

import pandas as pd

df = pd.read_csv('data.csv')

columns_to_drop =[]
for column in df.columns:
    if df[column].isnull().mean() > 0.25:
        columns_to_drop.append(column)



df = df.drop(columns=columns_to_drop, axis=1)
```
