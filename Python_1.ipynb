````python
brand_list = [
  {
      'brand_name': 'Apple',
      'price': 950,
      'ram_storage': 32,
      'hard_disk': 256,

   },

  {
      'brand_name': 'Google',
      'price': 930,
      'ram_storage': 42,
      'hard_disk': 259,
  } ,

  {
      'brand_name': 'MSFT',
      'price': 935,
      'ram_storage': 46,
      'hard_disk': 249,
  } ,
  {
      'brand_name': 'NVidia',
      'price': 835,
      'ram_storage': 16,
      'hard_disk': 129,
  } ,


]
# print(brand_list[0].get('brand_name'))
# print(brand_list[1].get('brand_name'))

print('Brand\t Price($)\t RAM(GB) \t Storage(GB)\n','----------------------------------------------------\n')

for brand in brand_list:
    print(brand.get('brand_name'),'\t',brand.get('price'),'\t\t',brand.get('ram_storage'),'\t\t',brand.get('hard_disk'))


print('----------------------------------------------------\n')


# To find the min/max based on a specific key, provide a 'key' argument
print('Min price brand:', min(brand_list, key=lambda x: x['price']))
print('Max price brand:', max(brand_list, key=lambda x: x['price']))


print(brand_list[0])
# print(brand_list.index('Apple'))
# print(brand_list.index('Google'))
# print(brand_list.index('MSFT'))
# print(brand_list.index('NVidia'))
````
